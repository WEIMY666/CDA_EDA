```python
data['vote'] = 1*(data.vote=='Clinton')
data['vote'].value_counts()
data['selfLR'] = data['selfLR'].map({'extLib':1,'Lib':2,'sliLib':3,'Mod':4,'sliCon':5,'Con':6,'extCon':7}).astype(int)
data['ClinLR'] = data['ClinLR'].map({'extLib':1,'Lib':2,'sliLib':3,'Mod':4,'sliCon':5,'Con':6,'extCon':7}).astype(int)
data['DoleLR'] = data['DoleLR'].map({'extLib':1,'Lib':2,'sliLib':3,'Mod':4,'sliCon':5,'Con':6,'extCon':7}).astype(int)
data['PID'] = data['PID'].map({'strDem':1,'weakDem':2,'indDem':3,'indind':4,'indRep':5,'weakRep':6,'strRep':7}).astype(int)
data['educ'] = data['educ'].map({'MS':1,'HSdrop':2,'HS':3,'Coll':4,'CCdeg':5,'BAdeg':6,'MAdeg':7}).astype(int)
```
